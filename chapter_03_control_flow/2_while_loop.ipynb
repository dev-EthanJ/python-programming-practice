### 03-2. while문 (while loop)

#### while문의 기본 구조

```python
# 열 번 찍어 안넘어가는 나무 없다
tree_hit = 0
while tree_hit < 10:
    tree_hit += 1
    print(f'나무를 {tree_hit}번 찍었습니다.')
print(f'나무를 {tree_hit}번 찍어 넘어갔습니다.')
```

```python
# while문 만들기
prompt = """
1. Add
2. Delete
3. List
4. Quit

Enter Number
:"""

number = 0
while number != 4:
    print(prompt)
    number = int(input())
```

#### while loop 강제로 빠져나가기 > `break`

```python
coffee = 10
money = 300
while money:
    print("커피가 한 잔 나왔습니다")
    coffee -= 1
    print("남은 커피는 %d잔 입니다." % coffee)
    if coffee == 0:
        print("남은 커피가 없습니다. 판매를 중지합니다.")
        break
```

#### while문의 처음으로 돌아가기 > `continue`

```python
a = 0
while a < 10:
    a += 1
    if a % 2 == 0: continue
    print(a)
```

```python
# 1부터 10까지 3의 배수를 뺀 나머지 출력
a = 0
while a < 10:
    a += 1
    if a % 3 == 0: continue
    else: print(a)
```
