#### 들여쓰기 (Indentation)

```python
# 들여쓰기 모든 문장에 반드시 수행
money = True
if money:
    print('택시를 ')
# print('타고 ') > IndentationError
    print('가라')
```

```python
# 들여쓰기 같은 깊이로 수행
if True:
    print('This')
    print('is')
        # print('true.') > IndentationError
```

#### `and`, `or`, `not`

```python
# 돈이 3000원 이상 있거나 카드가 있다면 택시를 타고 가고, 그렇지 않으면 걸어가라
money = 2000
card = True
if money >= 3000 or card == True:
    print("택시 타고 가기")
else:
    print("걸어가기")
```

#### 다양한 조건 판단 `elif`

```python
# 주머니에 돈이 있으면 택시를 타고가고, 돈은 없지만 카드가 있으면 버스를 타고 가고, 둘 다 없으면 걸어가라.
pocket = ['cellphone', 'paper']
if 'money' in pocket:
    print("택시 타고 가자")
elif 'card' in pocket:
    print("버스 타고 가자")
else:
    print("걸어가자")
```

#### 조건부 표현식
- `변수 = 참일때의_값 if 조건문 else 거짓일때의_값`

```python
score = 70
if score >= 60:
    message = "pass"
else:
    message = "fail"

message
```

```python
score = 50
message = "pass" if score >= 60 else "fail"
message
```
