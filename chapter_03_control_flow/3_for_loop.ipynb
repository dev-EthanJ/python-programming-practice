### 03-3. for문 (for loop)

#### for문 기본구조
```python
for 변수 in 리스트or튜플or문자열:
    수행문1
    수행문2
```

#### for문 예제

````python
# 1. 전형적인 for문
my_list = ['one', 'two', 'three']
for i in my_list:
    print(i)
````

````python
# 2. 다양한 for문의 사용
tuple_list = [(1, 2), (3, 4), (5, 6)]
for (first, last) in tuple_list:
    print(first + last)
````

````python
# 3. for문의 응용
score_list = [90, 25, 67, 45, 80]

num = 0
for score in score_list:
    num += 1
    if score >= 60:
        print("%d번 학생은 합격입니다." % num)
    else:
        print("%d번 학생은 불합격압니다." % num)
````

#### for문과 continue문

````python
marks = [90, 25, 67, 45, 80]

number = 0
for mark in marks:
    number += 1
    if mark < 60: continue
    print("%d번 학생은 합격했습니다." % number)
````

#### for문과 함께 자주 사용하는 `range` 함수

````python
a = range(10)
a
````

````python
a = range(1, 11)
a
````

````python
# range 함수 예시
add_sum = 0
for i in range(1, 11):
    add_sum += i

print(add_sum)
````

````python
marks = [90, 25, 67, 45, 80]

for index in range(len(marks)):
    if marks[index] < 60: continue
    else:
        print("%d번 학생은 합격했습니다." % (index + 1))
````

````python
# 1 ~ 100까지 합
total = 0
for i in range(1, 101):
    total += i
print(total)
````

````python
# for문과 range 함수를 이용한 구구단
for i in range(2, 10):
    for j in range(1, 10):
        print(f'{i:2} * {j:2} = {i * j:2}')
````

#### list comprehension 이용하기
- list 안에 for문 포함

````python
a = [1, 2, 3, 4]
result = []

for num in a:
    result.append(num * 3)

result
````

````python
# list comprehension 사용
a = [1, 2, 3, 4, 5]

result = [num * 3 for num in a]
result
````

````python
# list comprehension 안에서 if 사용
my_list = [1, 2, 3, 4, 5, 6]

result = [num * 2 for num in my_list if num % 2 == 0]
result
````

list comprehension 문법
- `[표현식 for 항목 in 반복_가능_객체 if 조건문]`

for문 중첩 가능
 ``` 
[표현식 for 항목1 in 반복_가능_객체1 if 조건1  
    for 항목2 in 반복_가능_객체2 if 조건2
    for 항목n in 반복_가능_객체n if 조건n] 
```


````python
result = [x * y for x in range(2, 10)
                for y in range(1, 10)]

print(result)
````
